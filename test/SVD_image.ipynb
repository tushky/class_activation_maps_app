```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open('cat.jpg').convert('RGB')
```

```python
image
```

```python
image = np.array(image)
```

```python
r_img = np.zeros_like(image)
for j in range(3):
    img = image[:,:,j]
    a, b, c = np.linalg.svd(img)
    r = np.zeros_like(img, dtype=np.float32)
    index = 80
    for i in range(index):
        v = np.reshape(a[:,i], (img.shape[0], 1))
        u = np.reshape(c[i,:], (1, img.shape[1]))
        r += v*u*b[i]
    r_img[:,:,j] = r
r_img = Image.fromarray(r_img)
r_img
```
