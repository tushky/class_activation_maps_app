```python
image = Image.open('horse2.jpg').convert('RGB')
```

```python
image = np.array(image)
```

```python
plt.imshow(img, cmap='gray')
```

```python
a, b, c = np.linalg.svd(img)
```

```python
print(f'a shape {a.shape} \nb shape {b.shape} \nc shape {c.shape}')
```

```python
r_img = np.zeros_like(image)
for j in range(3):
    r = np.zeros_like(img, dtype=np.float32)
    for i in range(50):
        v = np.reshape(a[:,i], (img.shape[0], 1))
        u = np.reshape(c[i,:], (1, img.shape[1]))
        r += v*u*b[i]
        #print(np.mean(r))
```
